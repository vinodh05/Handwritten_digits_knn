```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_excel('/content/ImageSegData_test.xls')
data=np.asarray(df)
classes=data[:,:1]
feature_vector=data[:,1:]
feature_vector[:,0]
labels=np.unique(classes)
classes_3=['BRICKFACE', 'CEMENT', 'FOLIAGE']
for i in classes_3:
    print("Class label ",i)
    x=[]
    y=[]
    count=0
    for j in range(classes.size):
        if i==data[j:j+1,:1]:
            x.append(data[j:j+1,1:2])
            y.append(data[j:j+1,2:3])
            count=count+1
    class_prob=count/classes.size
    x=np.array(x)
    x=np.reshape(x,-1)
    p_x_1=x/class_prob
    y=np.array(x)
    y=np.reshape(x,-1)
    p_y_1=y/class_prob
    plt.hist(p_x_1)
    plt.show()
    print("histogram for 2 features")
    plt.hist2d(p_x_1,p_y_1)
    plt.show()
```
